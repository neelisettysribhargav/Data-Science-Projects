```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./datasets/Dataset.csv')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Cuisines'].fillna(df['Cuisines'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Has Table booking'] = df['Has Table booking'].map({"Yes": 1, "No": 0})
df['Has Online delivery'] = df['Has Online delivery'].map({"Yes": 1, "No": 0})
df['Is delivering now'] = df['Is delivering now'].map({"Yes": 1, "No": 0})
df['Switch to order menu'] = df['Switch to order menu'].map({"Yes": 1, "No": 0})
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df["Aggregate rating"], bins=20, kde=True, color = 'skyblue')
plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
import folium
from folium.plugins import MarkerCluster

sample_df = df.head(100)
map_ = folium.Map(location=[sample_df["Latitude"].mean(), sample_df["Longitude"].mean()], zoom_start=10)
marker_cluster = MarkerCluster().add_to(map_)
for _, row in sample_df.iterrows():
    folium.Marker([row["Latitude"], row["Longitude"]], popup=row["Restaurant Name"]).add_to(marker_cluster)
map_
```

```python
country_distribution = df['Country Code'].value_counts()
city_distribution = df['City'].value_counts()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=country_distribution.index[:10], y=country_distribution.values[:10], color = 'skyblue')
plt.title("Top 10 Countries with Most Restaurants")
plt.xlabel("Country Code")
plt.ylabel("Number of Restaurants")
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x=city_distribution.index[:10], y=city_distribution.values[:10], color = 'skyblue')
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
location_features = df[['Latitude', 'Longitude']]
ratings = df['Aggregate rating']
correlation_values = location_features.corrwith(ratings)
print("Correlation between location and rating:\n", correlation_values)
```
