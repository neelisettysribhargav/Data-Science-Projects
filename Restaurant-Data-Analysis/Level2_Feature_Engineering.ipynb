```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./datasets/Dataset.csv')
```

TASK - 1

```python
table_booking_perc = df['Has Table booking'].value_counts(normalize=True) * 100
online_delivery_perc = df['Has Online delivery'].value_counts(normalize=True) * 100
print("Percentage of restaurants with Table Booking:",table_booking_perc)
print("Percentage of restaurants with Online Delivery:",online_delivery_perc)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="Has Table booking", y="Aggregate rating", data=df, color = 'skyblue')
plt.title("Comparison of Ratings: Table Booking vs No Table Booking")
plt.xlabel("Has Table Booking")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
colors = ['skyblue', 'red']
sns.countplot(x = "Price range", hue = "Has Online delivery", data = df, palette = colors)
plt.xlabel("Price Range")
plt.ylabel("Count of Resturants")
plt.title("Online Delivery Availability Across Price Ranges")
plt.legend(title = "Has Online Delivery")
plt.show()
```

TASK - 2

```python
most_common = df['Price range'].mode()[0]
print("Most common price range:", most_common)
```

```python
df.groupby('Price range')['Aggregate rating'].mean()
```

```python
sns.barplot(x = 'Price range', y ='Aggregate rating', data = df, color = 'green')
plt.title("Total ratings per Price range")
plt.show()
```
