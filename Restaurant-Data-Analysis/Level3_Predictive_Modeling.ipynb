```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./datasets/Dataset.csv')
```

TASK - 2

```python
cuisine_rating_pivot = df.groupby("Cuisines")["Aggregate rating"].mean().reset_index()
cuisine_rating_pivot = cuisine_rating_pivot.pivot_table(values="Aggregate rating", index="Cuisines")

plt.figure(figsize=(10, 5))
sns.heatmap(cuisine_rating_pivot.head(20), annot=True, cmap="plasma", fmt=".2f")
plt.title("Heatmap of Average Ratings Per Cuisine")
plt.show()
```

```python
cuisine_popularity = df.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False).head()
print("Most popular cuisines based on votes:",cuisine_popularity)
```

```python
high_rating_cuisines = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
print("Top rated Cuisines:",high_rating_cuisines.head())
```

TASK - 3

```python
sns.histplot(df["Aggregate rating"], bins=15, kde=True, color = 'skyblue')
plt.title("Distribution of Aggregate Rating")
plt.show()
```

```python
sns.boxplot(y=df["Aggregate rating"], color = 'skyblue')
plt.title("Distribution of Aggregate Rating")
plt.show()
```

```python
sns.violinplot(x=df["Aggregate rating"], color = 'skyblue')
plt.title("Distribution of Aggregate Rating")
plt.show()
```

```python
city_ratings_pivot = df.groupby('City')['Aggregate rating'].mean().reset_index()
city_ratings_pivot = city_ratings_pivot.pivot_table(values = 'Aggregate rating', index = 'City')
plt.figure(figsize = (10,20))
sns.heatmap(city_ratings_pivot, annot=True, cmap="magma", fmt=".2f")
plt.title("City va Aggregate Rating")
plt.xlabel("City")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
df1 = df[['Price range', 'Votes', 'Average Cost for two', 'Latitude', 'Longitude']]
y = df['Aggregate rating']
correlation_values = df1.corrwith(y)
correlation_df = correlation_values.to_frame()
correlation_df.columns = ['Aggregate rating']
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_df, annot=True, cmap='inferno', fmt=".2f")
plt.title("Relationship between Features and Aggregate Rating")
plt.show()
```
